```python
import math
```

```python
import pandas as pd
from sqlalchemy import create_engine 
 
# DB 커넥션 열기
engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl') 

# DB 테이블을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_sql_query('SELECT REGIONID, PRODUCTGROUP, YEARWEEK, VOLUME FROM kopo_product_volume', engine) 
```

```python
customerData
```

```python
# 컬럼명 대문자
customerData.columns = [x.upper() for x in customerData.columns]
```

```python
# 정렬하기
customerData=customerData.sort_values(["REGIONID","PRODUCTGROUP","YEARWEEK"], ascending=[True,True,True])
# 인덱스 다시 설정
sortedData = customerData.reset_index(drop=True)
```

```python
def sub_function(data):
    data = data.reset_index(drop=True)
    data["MA"]=data["VOLUME"].rolling(window=5,center=True).mean()
    
    suborder=math.floor(5/2)
   
    list=[]
    maxLength=len(data)-1
    for i in range(0,suborder):
        list.append(data['VOLUME'][0:i+suborder+1].mean())
        data.loc[i,"MA"]=list[i]
    
    list1=[]
    for i in range(0,suborder):
        list1.append(data['VOLUME'][-i-suborder-1:].mean())
        data['MA'][maxLength-i]=list1[i]
        
    return data
```

```python
groupResult11 = sortedData.groupby(['REGIONID','PRODUCTGROUP']).apply(sub_function)
groupResult11
```
