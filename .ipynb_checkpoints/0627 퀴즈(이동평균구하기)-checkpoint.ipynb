```python
import pandas as pd
from sqlalchemy import create_engine 
 
# DB 커넥션 열기
engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl') 

# DB 테이블을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_sql_query('SELECT REGIONID, PRODUCTGROUP, YEARWEEK, VOLUME FROM kopo_product_volume', engine) 
```

```python
customerData.head(5)
```

```python
# 컬럼명 대문자
customerData.columns = [x.upper() for x in customerData.columns]
```

```python
# 정렬하기
customerData=customerData.sort_values(["REGIONID","PRODUCTGROUP","YEARWEEK"], ascending=[True,True,True])
customerData
```

```python
# 인덱스 다시 설정
sortedData = customerData.reset_index(drop=True)
```

```python
sortedData.head(5)
```

```python
# 이동평균 구하기
sortedData["TEST"]=sortedData['VOLUME'].rolling(window=5, center=True).mean()
sortedData.head(5)
```

```python
import math
suborder=math.floor(5/2)
suborder
```

```python
# 첫 두 행 이동평균 구하기
list=[]
for i in range(0,suborder):
    list.append(sortedData['VOLUME'][0:i+suborder+1].mean())
list
```

```python
for i in range(0,suborder):
    sortedData.loc[i,"TEST"]=list[i]
```

```python
# 마지막 두행 이동평균 구하기
list1=[]
num=len(sortedData)-1
for i in range(0,suborder):
    list1.append(sortedData['VOLUME'][-i-suborder-1:].mean())
list1
```

```python
for i in range(0,suborder):
    sortedData['TEST'][num-i]=list1[i]
```
