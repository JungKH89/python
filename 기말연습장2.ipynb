```python
import pandas as pd
from sqlalchemy import create_engine 
import requests, bs4, pandas as pd, numpy as np
```

```python
resp = requests.get('https://www.timeanddate.com/holidays/south-korea/')
resp.encoding='utf-8'
html=resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')
```

```python
columns=bs.select(' tr.head > th.sep' )
columns
```

```python
columnslist1=["Date"]
columnslist1
```

```python
contents1=bs.select(' tbody > tr.c0 > th.nw')
contents1
```

```python
contents=bs.select(' tbody > tr ')
contents
```

```python
columnlist=[]
for column in columns:
    columnlist.append(column.text)
columnlist
```

```python
lists=[]
alldfcontents=[]
lists1=[]
datelists=[]
```

```python
for content in contents1:
    if(content.text=="None"):
         continue;
    lists1.append(content.text)
    datelists.append(lists1)
    lists1=[]
    
datelists
```

```python
for content in contents:
    tds=content.findAll("td")
    for num in tds:
        if(num.text=="None"):
            continue;
        lists.append(num.text)
    alldfcontents.append(lists)
    lists=[]
alldfcontents
```

```python
result=pd.DataFrame(columns=columnlist, data=alldfcontents )
result2=pd.DataFrame(columns=columnslist1, data=datelists )
result2
```

```python
result.columns = ["WEEK","HOLIDAY_NAME","HOLIDAY_TYPE"]
result2.columns = ["DATE"]
result
```

```python
finalresult = pd.concat([result2,result], axis=1)
finalresult
```
